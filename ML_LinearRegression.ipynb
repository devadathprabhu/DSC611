```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([2,4,5])
Y = np.array([1.2,2.8,5.3])
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
b0 = 0              #intercept
b1 = 1              #slope
alpha = 0.001       #learning rate
iterations = 1000   #number of iterations
error = []          #record error in each iteration
```

```python
for itr in range(iterations):
  error_cost = 0
  cost_b0 = 0
  cost_b1 = 0
  for i in range(len(X)):
    y_pred = (b0 + b1 * X[i])
    error_cost = error_cost + (Y[i] - y_pred)**2

    for j in range(len(X)):
      partial_b0 = -1 * (Y[j] - (b0 + b1 * X[j]))           # (b0 + b1 * X[j]) is predicted y
      partial_b1 = -1 * X[j] * (Y[j] - (b0 + b1 * X[j]))

      cost_b0 = cost_b0 + partial_b0
      cost_b1 = cost_b1 + partial_b1

  b0 = b0 + alpha * cost_b0
  b1 = b1 + alpha * cost_b1

  error.append(error_cost/(len(X)*2))
```
